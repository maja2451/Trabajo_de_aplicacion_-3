# datos de entrenamiento

```python
# Resultados de ambos modelos
resultados = {
    "Modelo 1": {"RMSE": 136.1732, "MAE": 13.74, "R2": -2.94,"MSE":300.14}, 
    "Modelo 2": {"RMSE": 8.65, "MAE": 7.48, "R2": 0.02,"MSE":74.86},
    "Modelo 3": {"RMSE": 0.00, "MAE": 0.00, "R2": 1.00,"MSE":0.00},
}

# Comparación de métricas
for modelo, metricas in resultados.items():
    print(f"Resultados para {modelo}:")
    for metrica, valor in metricas.items():
        print(f"  {metrica}: {valor}")
    print()

# Determinar el modelo con mejor RMSE
mejor_modelo = min(resultados, key=lambda x: resultados[x]["RMSE"])
print(f"El modelo con mejor RMSE es: {mejor_modelo}")

# Visualización de las métricas
import matplotlib.pyplot as plt

metricas = ["RMSE", "MAE", "R2"]
valores_modelo1 = [resultados["Modelo 1"][m] for m in metricas]
valores_modelo2 = [resultados["Modelo 2"][m] for m in metricas]

x = range(len(metricas))
plt.bar(x, valores_modelo1, width=0.4, label="Modelo 1", align="center")
plt.bar([p + 0.4 for p in x], valores_modelo2, width=0.4, label="Modelo 2", align="center")
plt.xticks([p + 0.2 for p in x], metricas)
plt.ylabel("Valor")
plt.title("Comparación de métricas entre modelos")
plt.legend()
plt.show()
```
